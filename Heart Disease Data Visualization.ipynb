### `SleepTime` and `BMI`
Create a plot with three subplots in the same layout of Section 3.3.1, Part 4 More Complicated Arrangements. The main plot should be a scatterplot with `Sleeptime` and `BMI`. The two smaller plots show the histograms of these two variables.



```python



fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.4, wspace=0.4)

main_ax = fig.add_subplot(grid[:-1, 1:])

plt.title(' Sleep Time Versus BMI')

y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

y= Heart2020['BMI']
x= Heart2020['SleepTime']

main_ax.plot(x, y, 'ok', color = 'blue', markersize= 2)

plt.xlabel('SleepTime')


x_hist.hist(x, 40, histtype='stepfilled', orientation='vertical', color='orange')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',orientation='horizontal', color='orange')
plt.ylabel('BMI')

y_hist.invert_xaxis()

```

### 2-Dimensional Binning
Create a 2-dimensional hexagonal binplot  to show the relationship between `SleepTime` and `BMI`.


```python
fig = plt.figure(figsize=(8,6))
plt.hexbin(x, y, gridsize=15, cmap='Purples')
cb = plt.colorbar(label='counts in bin')
```
